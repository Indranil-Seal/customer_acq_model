```python
def one_hot_encode_and_sparsity(df, column_name):
    """
    One-hot encodes a categorical column and calculates the percentage increase in sparsity.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to one-hot encode.
        
    Returns:
        pd.DataFrame: DataFrame with one-hot encoded features.
        float: Percentage increase in sparsity.
    """
    # Original number of non-zero elements
    original_non_zero = np.count_nonzero(df[column_name].notna())

    # Perform one-hot encoding
    df_encoded = pd.get_dummies(df, columns=[column_name], dtype=int)

    # New number of non-zero elements
    new_non_zero = np.count_nonzero(df_encoded.values)

    # Calculate sparsity increase percentage
    sparsity_increase = ((new_non_zero - original_non_zero) / original_non_zero) * 100

    return df_encoded, sparsity_increase

# Example usage:
# df = pd.DataFrame({'category': ['A', 'B', 'C', 'A', 'B', 'C']})
# encoded_df, sparsity_increase = one_hot_encode_and_sparsity(df, 'category')
# print(encoded_df)
# print(f"Sparsity Increase: {sparsity_increase:.2f}%")
```

```python
def map_food_department(df, column_name):
    """
    Maps specific food department categories to numerical labels and keeps the rest unchanged.

    Parameters:
        df (pd.DataFrame): Input DataFrame containing the food_department column.
        column_name (str): The name of the column to map.

    Returns:
        pd.DataFrame: DataFrame with a new column containing mapped values.
    """
    # Define the mapping for selected categories
    mapping = {
        'Snack Foods': 1,
        'Produce': 2,
        'Dairy': 3,
        'Baking Goods': 4,
        'Deli': 5
    }

    # Apply mapping, keeping other values unchanged
    #df['food_department_mapped'] = df[column_name].map(mapping).fillna(df[column_name])
    df['food_department_mapped'] = df[column_name].map(mapping).fillna(0).astype(int)

    return df

# Example usage:
# df = pd.DataFrame({'food_department': ['Snack Foods', 'Produce', 'Dairy', 'Meat', 'Beverages', 'Baking Goods']})
# df = map_food_department(df, 'food_department')
# print(df)
```
