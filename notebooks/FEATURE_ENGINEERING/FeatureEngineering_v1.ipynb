```python
def map_food_department(df, column_name):
    """
    Maps specific food department categories to numerical labels and keeps the rest unchanged.

    Parameters:
        df (pd.DataFrame): Input DataFrame containing the food_department column.
        column_name (str): The name of the column to map.

    Returns:
        pd.DataFrame: DataFrame with a new column containing mapped values.
    """
    # Define the mapping for selected categories
    mapping = {
        'Snack Foods': 1,
        'Produce': 2,
        'Dairy': 3,
        'Baking Goods': 4,
        'Deli': 5
    }

    # Apply mapping, keeping other values unchanged
    #df['food_department_mapped'] = df[column_name].map(mapping).fillna(df[column_name])
    df['food_department_mapped'] = df[column_name].map(mapping).fillna(0).astype(int)

    return df

# Example usage:
# df = pd.DataFrame({'food_department': ['Snack Foods', 'Produce', 'Dairy', 'Meat', 'Beverages', 'Baking Goods']})
# df = map_food_department(df, 'food_department')
# print(df)
```

```python
def one_hot_encode_food_categories(df, food_column, top_categories):
    
    for category in top_categories:
        df['category'] = (df[food_column] == category).astype(int)
        
    return df
```

```python
def create_dummy_variables(data, column_name):
    if column_name not in data.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFame")

    dummy_df = pd.get_dummies(data[column_name], prefix=column_name, drop_first=False)
    dummy_df = dummy_df.map(lambda x:1 if x is True else 0)
    result_df = pd.concat([data, dummy_df], axis=1).drop(columns=[column_name])

    return result_df
```

```python
data_ = create_dummy_variables(data_, 'promotion_name')
data_
```

## Feature Engineering: avg. yearly_income

```python
def extract_income_range(data, column_name):
    if column_name not in data.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFrame")

    data[['Min_Income', 'Max_income']] = data[column_name].str.extract(r'\$(\d+)K - \$(\d+)K')
    data[['Min_Income', 'Max_income']] = data[['Min_Income', 'Max_income']].astype(float) * 1000

    return data
```

```python
data_ = extract_income_range(data_, 'avg. yearly_income')
data_
```
